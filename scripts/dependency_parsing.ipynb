```python
#Importar librerías
```

```python
import re
import os
import spacy
from collections import Counter
import ast
import pandas as pd
nlp=spacy.load("es_core_news_sm")
```

```python
#Abrir cedulario con codificación utf-8
```

```python
with open("cedulario.txt", "r", encoding="utf-8") as f:
  texto=f.read()
```

```python
#Importar diccionario y definir función para normalizar las entidades
```

```python
 diccionario={"concha":"conchos","indios":"indios"...}
```

```python
def normalizar(entidad):
  entidad=entidad.lower()
  return diccionario.get(entidad, entidad)
```

```python
#Procesar texto y definir los grupos para el dependency parsing
```

```python
doc=nlp(texto)
grupos=["conchos", "tarahumaras", "tobosos", "pimas", "indios"]
```

```python
#Crear variable edges para almacenar y extrar verbos (VERB), adjetivos (ADJ), y advervios (ADV)
```

```python
edges=[]
for token in doc:
     grupo=normalizar(token.text)
     if grupo in grupos:
             for child in token.children:
                     if child.pos_=='ADJ' or child.pos_=='VERB':
                             edges.append((grupo, child.text))
```

```python
#Realizar conteo e importar resultados en formato .csv para Gephi
```

```python
conteo=Counter(edges)
df_edges=pd.DataFrame([(source, target, weight) for (source, target), weight in conteo.items()], columns=["source", "target", "weight"])
df_edges.to_csv("edges_pos.csv", index=False, encoding="utf-8")
```
