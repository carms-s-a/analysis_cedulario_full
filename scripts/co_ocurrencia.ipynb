```python
# Importar librerias
```

```python
import pandas as pd
import itertools
from collections import defaultdict
import re
import os
from pathlib import Path
```

```python
# Importar el diccionario de normalización
```

```python
 diccionario={"concha":"conchos","indios":"indios"...}
```

```python
# importar el texto y dividir cédulas con "[FIN]"
```

```python
with open("utblac_wbs_0496_pt1.txt", "r", encoding="utf-8") as f:
  texto = f.read()
cedulas = texto.split("[FIN]")
```

```python
# Crear función de normalización
```

```python
def normalizar(entidad):
    entidad = entidad.lower()
    return diccionario.get(entidad, entidad)
```

```python
# Lista de entidades
```

```python
entidades = [
  "real hacienda",
"francisco fernández de la cueva y enríquez de cabrera"...
]
```

```python
# Crear función para simluación de NER por cédula
```

```python
def extraer_entidades(cedula, lista_entidades):
    cedula = cedula.lower()
    return [e for e in lista_entidades if e in cedula]
```

```python
# Extracción de co-ocurrencias
```

```python
edges = defaultdict(int)
for cedula in cedulas:
     ents = extraer_entidades(cedula, entidades)
     ents_norm = [normalizar(e) for e in ents]
     ents_norm = list(set(ents_norm))
     for a, b in itertools.combinations(ents_norm, 2):
             edge = tuple(sorted([a, b]))
             edges[edge] += 1
rows = [(a, b, w) for (a, b), w in edges.items()]
```

```python
# Importar a formato .csv para Gephi
```

```python
edges_df = pd.DataFrame(rows, columns=["source", "target", "weight"])
edges_df.to_csv("edges_gephi.csv", index=False, encoding="utf-8")
```
