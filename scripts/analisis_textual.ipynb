#División de cédulas


```
# Importar librerías
```

```
import spacy
import pandas as pd
import re
import os
import csv
nlp = spacy.load("es_core_news_sm")
```

```
# Leer el archivo
```

```
with open("utblac_wbs_0496.txt", "r", encoding="utf-8") as f:
  texto = f.read()
```

```
# Dividir cédulas usando [FIN] como divisor
```

```
cedulas = texto.split("[FIN]")
```

```
# Crear la carpeta para guardar los resultados
```

```
os.makedirs("cedulas", exist_ok=True)
```

```
# Guardar cédulas como archivos .txt
```

```
for i, c in enumerate(cedulas):
  with open(f"cedulas/cedula_{i+1}.txt", "w", encoding="utf-8") as out:
    out.write(c.strip())
```

# Named Entity Recognition (NER)

```
# Definir función para normalizar
```

```
def normalize(text):
	text=str(text)
	text=text.lower()
	text=text.strip()
	text = re.sub(r"\s+", " ", text)
	text = re.sub(r"[^\w\sáéíóúñü]", "", text)
	return text
```
